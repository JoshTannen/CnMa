### Filter those with hashes and commands

```python
df_hash = df_os[(df_os.hashes!="['']")&(df_os.hashes!="[]")]
```

```python
df_hash
```

```python
#count of hashes stored as dic key pair
#count of commands, stored as dic key pair. 
hashdic = {}

for i,row in tqdm(df_hash.iterrows()):
    for c in ast.literal_eval(row['hashes']):
        if c in hashdic.keys():
            hashdic[c]+=1
        else:
            hashdic[c]=1
hashdic
```

```python
df_hash['malfam'] = None

for i,row in tqdm(df_hash.iterrows()):
    malfam = []
    for h in ast.literal_eval(row['hashes']):
        #print(h)
        for j,ro in exp.iterrows():
            #print(ro.hash)
            if h == ro.hash:
                malfam.append(ro.classification)
    df_hash.at[i,"malfam"]=malfam

df_hash        
```
