# TTP tagging on commands 
#### (based on GroupTracer paper)

```python
#import IP_address_filter
#import SG_IP_matching
#import xml_to_csv
import VT_API
import OS_API
import pandas as pd
import numpy as np
from datetime import datetime
import time
import os
from tqdm import tqdm
import json
import re
import csv
import ast
import matplotlib.pyplot as plt
```

```python
#columns of interest
COLS = ['peerIP',
        'peerPort',
        'hostIP',
        'hostPort',
        'commands',
        'hashes',
        'urls',
        'loggedin',
        'startTime',
        'endTime',
        'sort_num',
        'peerCountry',
        'hostCountry'
       ]

SIZE = 10000
```

```python
defaultquery = [{"range": {
                "startTime": {
                  "gte": "now-1d",
                  "lte": "now",
                  "format": "strict_date_optional_time"
                }}},
                {"exists": {"field": "commands"}}
               ]
```

```python
size = 10000 #batch size, max 10000
search_after = [0]
df_os = pd.DataFrame(columns=COLS)

queryls = defaultquery
print(queryls)

total = 0

while (size == SIZE):
    print(f"Continue from record: {search_after}")
    data = OS_API.opensearch_request(size,queryls,search_after)
    df_os = pd.concat([df_os,data])
    
    #update params
    size = len(data) #exit once it completes last batch
    total+=size #update total records number
    print(f'Current total: {total}')
    search_after = data['sort_num'][len(data)-1] #update sort number to continue after
    #time.sleep(1)
    
print(f'Done. Total records: {total}')
```

```python
#pull http or urls from commands
commandls = []
index = []

for i,row in df_os.iterrows():
    for c in row['commands']:
        try:
            #regex to find url starting with http(s)
            found = re.findall("https?:\\/\\/(?:www\\.)?[-a-zA-Z0-9@:%._\\+~#=]{1,256}\\.[a-zA-Z0-9()]{1,6}\\b(?:[/]*)", c)
            #print(i, found)

            for f in found:
                if f not in commandls:
                    commandls.append(f)

        except AttributeError:
            print(i, 'No matches found')
            found = ''
```

### List of Unique commands

```python
#count of commands, stored as dic key pair. 
cmdic = {}

for i,row in df_os.iterrows():
    for c in ast.literal_eval(row['commands']):
        if c in cmdic.keys():
            cmdic[c]+=1
        else:
            cmdic[c]=1
cmdic
```

```python
len(cmdic)
```

```python
#sort descending by count
cmdicsort = {k: v for k, v in sorted(cmdic.items(), key=lambda item: item[1], reverse = True)}
```

```python
#plot top 10 commands count
commands = list(cmdicsort.keys())[:14]
counts = list(cmdicsort.values())[:14]

fig = plt.figure(figsize = (10, 5))
plt.barh(commands, counts, color ='navy')
```

```python
totalcount = sum(cmdicsort.values())
print(totalcount)

#plot top 10 commands percentage
ratio = [x/totalcount for x in counts]

fig = plt.figure(figsize = (10, 5))
plt.barh(commands, ratio, color ='orange')
```

```python
kbase = pd.read_csv('knowledgebase.csv')
kbase
```

```python
#assign ttps to commands
cmdf = pd.DataFrame.from_dict({"commands": list(cmdicsort.keys()), "count": list(cmdicsort.values())})
#cmdf

cmdf['ttps'] = None
kls = kbase['command']
#counter = 0

for i,row in tqdm(cmdf.iterrows()):
    #create exception for the "w" command
    if row['commands']!="w":
        ttps = [] #blank ttp list

        for k in kls:
            if k in row['commands']:
                ttp = list(kbase.loc[kls == k,['tactic','technique','id']].values[0])
                if ttp not in ttps:
                    ttps.append(ttp)

        cmdf.at[i,'ttps'] = ttps
    
    else:
        cmdf.at[i,'ttps'] = [["Discovery","System Owner/User Discovery","T1033"]]
        continue

cmdf
```
